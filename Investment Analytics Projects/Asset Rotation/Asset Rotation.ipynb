```python
class SBBIDataProcessor:
    def __init__(self, file_path, sheet_name):
        self.file_path =  file_path
        self.sheet_name  =  sheet_name 
        self.data = None 
        self.column_rename_map = {
            "More Descriptive Name of Series:": "Date",
            "Ibbotson® SBBI® US Large-Cap Stocks (Total Return)": "Large Cap Total Return",
            "Ibbotson® SBBI® US Large-Cap Stocks (Income Return)": "Large Cap Income Return",
            "Ibbotson® SBBI® US Large-Cap Stocks (Capital Appreciation  Return)": "Large Cap Capital Return",
            "Ibbotson® SBBI® US Small-Cap Stocks (Total Return)": "Small Cap Total Return",
            "Ibbotson® SBBI® US Long-term (20-Year) Corporate Bonds (Total Return)": "Corp Bonds Total Return",
            "Ibbotson® SBBI® US Long-term (20-Year) Government Bonds (Total Return)": "Gov Bonds Total Return",
            "Ibbotson® SBBI® US Long-term (20-Year) Government Bonds (Income Return)": "Gov Bonds Income Return",
            "Ibbotson® SBBI® US Long-term (20-Year) Government Bonds (Capital Appreciation Return)": "Gov Bonds Capital Return",
            "Ibbotson® SBBI® US Long-term (20-Year)  Government Bonds (Yield)": "Gov Bonds Yield",
            "Ibbotson® SBBI® US Intermediate-term (5-Year) Government Bonds (Total Return)": "Int Gov Bonds Total Return",
            "Ibbotson® SBBI® US Intermediate-term (5-Year) Government Bonds (Income Return)": "Int Gov Bonds Income Return",
            "Ibbotson® SBBI® US Intermediate-term (5-Year) Government Bonds (Capital Appreciation Return)": "Int Gov Bonds Capital Return",
            "Ibbotson® SBBI® US Intermediate-term (5-Year) Government Bonds (Yield)": "Int Gov Bonds Yield",
            "Ibbotson® SBBI® US (30-Day) Treasury Bills": "30-Day TBill Total Return",
            "Ibbotson® SBBI® US Inflation": "Inflation Rate"
        }
    def load_data(self):
        self.data = pd.read_excel(file_path, sheet_name = self.sheet_name, skiprows = 10)

    def rename_columns(self):
        if self.data is not None:
            self.data.rename(columns = self.column_rename_map, inplace = True)
        else: raise ValueError ('Data is not called.')
    
    def filter(self, columns_to_keep):
        if self.data is not None:
            self.data = self.data[columns_to_keep]
        else: raise ValueError('Data is not called')

    def show_data(self):
        return self.data
```
